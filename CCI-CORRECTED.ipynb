#### Import Data

```python
import pandas as pd
dat= pd.read_csv('data_jun.csv', header=[0,1], index_col=[0])
```

### CCI Calculation

```python
#dat=dat.iloc[1201:].fillna(method='ffill')
TYP= (dat['Close']+dat['Low']+dat['High'])/3
```

```python
PTmean = TYP.rolling(window = 20, min_periods=10).mean()#.fillna(value=0)

```

```python
import numpy as np
mads = lambda x: np.fabs(x - x.mean()).mean()
Mad = TYP.fillna(method='ffill').rolling(window=20).apply(mads, raw=True)
```

```python
CCI05=(TYP - PTmean)/(Mad*.015)
CCI05
#https://www.profitspi.com/stock/view.aspx?v=stock-chart&uv=100564&p=LUV
```

```python
Ress = CCI05.tail(1).T
Ress.to_csv('cci_jul.csv')
```

### Simple Moving Average

```python
dat['Close'].rolling(window=336, min_periods=19).mean().tail(1).T.to_csv("336DAYS.csv")
dat['Close'].rolling(window=1460, min_periods=19).mean().tail(1).T.to_csv("1460DAYS.csv")
dat['Close'].rolling(window=200, min_periods=19).mean().tail(1).T.to_csv("200DAYS.csv")
dat['Close'].rolling(window=50, min_periods=19).mean().tail(1).T.to_csv("50DAYS.csv")
dat['Close'].rolling(window=10, min_periods=1).mean().tail(1).T.to_csv("10DAYS.csv")
```
