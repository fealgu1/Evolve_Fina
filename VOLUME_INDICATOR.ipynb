### Function OBV

```python
def OBV(DF):
    df =data.copy()
    delta = df['Close'].pct_change()
    dire=np.where(delta>=0,1,-1)
    dire[0]=0
    voldic = dire*df['Volume']
    obv = voldic.cumsum()
    return obv
    
```

```python
DF = OBV(data)

#df = data.copy()
DF
```

```python
import yfinance as yf
import numpy as np
import datetime as dt
import statsmodels.api as sm

def bvolpe(DF, n):
    df = DF.copy()
    norm = (df-df.min())/(df.max()-df.min())
    x = np.array(range(len(norm)))
    x = (x - x.min())/(x.max() - x.min())
    slopes = [i*0 for i in range(n-1)]
    for i in range(n,len(norm)+1):
        y_scaled = norm[i-n:i]
        x_scaled = x[:n]
        x_scaled = sm.add_constant(x_scaled)
        model = sm.OLS(y_scaled,x_scaled)
        results = model.fit()
        slopes.append(results.params)
    #slope_angle = (np.rad2deg(np.arctan(np.array(slopes))))
    slope_angle = (np.rad2deg(np.arctan((slopes)[-1])))
    #dire=np.where(results>=0,"Up","Down")
    return (slope_angle) #np.array
    
```

```python
final = bvolpe(DF,5).tail()
final.columns = DF.columns
final
final.loc['x1',:].to_csv('BVO_As200.csv')
```
