## CCI

```python
import pandas as pd
import numpy as np

def calculate_cci(clean_data):
    # Calculate Typical Price (TYP)
    TYP = (clean_data['Close'] + clean_data['Low'] + clean_data['High']) / 3
    
    # Calculate the rolling mean of TYP
    PTmean = TYP.rolling(window=20, min_periods=10).mean()
    
    # Function to calculate Mean Absolute Deviation (MAD)
    mads = lambda x: np.fabs(x - x.mean()).mean()
    
    # Calculate the rolling MAD
    Mad = TYP.fillna(method='ffill').rolling(window=20).apply(mads, raw=True)
    
    # Calculate the Commodity Channel Index (CCI)
    CCI05 = (TYP - PTmean) / (Mad * 0.015)
    
    return CCI05

# Example usage:
# clean_data = pd.read_csv('your_data.csv') # Make sure your DataFrame has 'Close', 'Low', and 'High' columns
cci_values = calculate_cci(clean_data)

```

```python
cci = cci_values.stack(level =0).reset_index()


cci = cci.rename(columns={'level_1':'symbol'})

# reshape the dataframe
cci_tb = cci.melt(id_vars = ['Date','symbol'], var_name = 'indicator', value_name = 'Value')
#meltrsi_3 = meltrsi_3.dropna()

# add a value name
cci_tb['indicator']= 'CCI'
```

## OBV

```python
def OBV(DF):
    df =clean_data.copy()
    delta = df['Close'].pct_change()
    dire=np.where(delta>=0,1,-1)
    dire[0]=0
    voldic = dire*df['Volume']
    obv = voldic.cumsum()
    return obv
    
```

```python
obv = OBV(clean_data)
```

```python
import pandas as pd
import psycopg2

# Load the data from the CSV
# df = pd.read_csv('/mnt/data/data.csv')  # Adjust the file path accordingly
# Assuming 'data' is your DataFrame

# Convert 'Date' column to datetime format
nasdaq['date'] = pd.to_datetime(nasdaq['date'])
df = nasdaq.copy()

# Connect to the PostgreSQL database
conn = psycopg2.connect(
    dbname='evolve_db', 
    user='postgres', 
    password='ingrain', 
    host='localhost', 
    port='5432'
)
cur = conn.cursor()

# SQL to create the tables
create_tickers_table = """
CREATE TABLE IF NOT EXISTS tickers (
    id SERIAL PRIMARY KEY,
    symbol_id VARCHAR(20) NOT NULL UNIQUE  -- Increased length for symbol_id
);
"""
create_stock_metrics_table = """
CREATE TABLE IF NOT EXISTS stock_metrics (
    id SERIAL PRIMARY KEY,
    date DATE NOT NULL,
    ticker_id INT NOT NULL,
    indicator VARCHAR(20) NOT NULL,  -- Increased length for indicator
    value NUMERIC(20, 10) NOT NULL,
    FOREIGN KEY (ticker_id) REFERENCES tickers(id),
    CONSTRAINT unique_record UNIQUE (date, ticker_id, indicator)
);
"""

cur.execute(create_tickers_table)
cur.execute(create_stock_metrics_table)
conn.commit()

# Insert unique tickers
tickers = df['symbol_id'].unique()
for ticker in tickers:
    cur.execute("INSERT INTO tickers (symbol_id) VALUES (%s) ON CONFLICT (symbol_id) DO NOTHING;", (ticker,))

conn.commit()

# Insert data into the stock_metrics table
for _, row in df.iterrows():
    cur.execute("""
        INSERT INTO stock_metrics (date, ticker_id, indicator, value)
        VALUES (
            %s,
            (SELECT id FROM tickers WHERE symbol_id = %s),
            %s,
            %s
        ) ON CONFLICT (date, ticker_id, indicator) DO NOTHING;
    """, (row['date'], row['symbol_id'], row['indicator'], row['value']))

conn.commit()
cur.close()
conn.close()
```

### Check for Duplicates

```python
import psycopg2
import pandas as pd

# Connect to the PostgreSQL database
conn = psycopg2.connect(
    dbname='evolve_db',
    user='postgres',
    password='ingrain',
    host='localhost',
    port='5432'
)
cur = conn.cursor()

# SQL query to find duplicates
find_duplicates_query = """
SELECT date, ticker_id, indicator, COUNT(*)
FROM stock_metrics
GROUP BY date, ticker_id, indicator
HAVING COUNT(*) > 1;
"""

cur.execute(find_duplicates_query)
duplicates = cur.fetchall()

# Convert the result to a DataFrame for better visualization
duplicates_df = pd.DataFrame(duplicates, columns=['date', 'ticker_id', 'indicator', 'count'])

print(duplicates_df)

cur.close()
conn.close()
```
