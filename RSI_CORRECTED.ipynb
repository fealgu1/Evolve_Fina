### Pull the data from yahoo website

```python
#RSI.to_csv('data05.csv')
RSI=pd.read_csv('data_jun.csv', header=[0,1], index_col=[0])
RSI
```

```python



Delta= RSI['Close'].diff()
Delta = Delta[1:]


```

```python

df = Delta.copy()
Delta[Delta < 0] = 0

```

```python
Gain = Delta
```

```python
Loss = df[df<0]

Loss[Loss<0]=abs(Loss[Loss<0])
Loss=Loss.fillna(value=0)
```

```python
exmax = Gain.ewm(com=4, adjust = False).mean()
exmin = Loss.ewm(com=4, adjust = False).mean()
```

```python
rs = exmax/exmin
```

```python
rs.index
```

```python
rsi = 100 - (100/(1+rs))
rsi.loc['2023-03-03 00:00:00'].to_csv('RSI-5.csv')
rsi['HA'].tail(1)
```
