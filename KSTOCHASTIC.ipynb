### 52 Weeks Data

```python
sh = KSTOCK['High'].tail(253)
maxdata = pd.concat([sh.max(), sh.idxmax()], axis=1)
maxdata.to_csv('hi_decs.csv')

s = KSTOCK['Low'].tail(253)
mindata = pd.concat([s.min(), s.idxmin()], axis=1)
mindata.to_csv('Lo_decs.csv')
```

### Rolling Close

```python
High = round(KSTOCK['Close'].rolling(window =253, min_periods=1).max(),2)#.dropna()
Low = round(KSTOCK['Close'].rolling(window =253, min_periods=1).min(),2)#.dropna()
#Low.to_csv('Loww.csv') 

hipctchg= High.pct_change(fill_method='ffill')
lopctchg= Low.pct_change(fill_method='ffill')
#test[lambda x: x.index != 0]
result = pd.concat([High, hipctchg, Low, lopctchg], axis=1, ignore_index=False)


col = result.columns
col = list(col)
col.sort()
dflow = result[col]
```
