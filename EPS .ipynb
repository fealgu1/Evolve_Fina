```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_yahoo_finance_tables(tickers):
    all_data = []
    headers = None  # Initialize headers
    retrieved_tickers = []  # Keep track of tickers for which data is successfully retrieved

    for ticker in tickers:
        url = f"https://finance.yahoo.com/quote/{ticker}/analysis"

        try:
            # Send a GET request to the URL with a specific user agent
            headerss = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
            response = requests.get(url, headers=headerss)

            # Check if the request was successful (status code 200)
            if response.status_code == 200:
                # Parse the HTML content of the page
                soup = BeautifulSoup(response.content, "html.parser")

                # Find the section with the specified data-testid attribute
                section = soup.find("section", {"data-testid": "epsTrend"})

                if section:
                    # Find the table within the section
                    table = section.find("table", class_="svelte-17yshpm")

                    # Extract column headers if not already extracted
                    if headers is None:
                        headers = [header.text.strip() for header in table.find("tr", class_="svelte-17yshpm").find_all("th")]

                    # Extract rows data
                    rows_data = []
                    rows = table.find_all("tr")
                    for row in rows:  # Iterate over all rows
                        row_data = [data.text.strip() for data in row.find_all("td")]
                        rows_data.append(row_data)

                    # Pad rows with None values to match the length of the longest row
                    max_row_length = max(len(row) for row in rows_data)
                    for row in rows_data:
                        if len(row) < max_row_length:
                            row.extend([None] * (max_row_length - len(row)))

                    # Append data to all_data list
                    all_data.extend(rows_data)

                    # Record the ticker for which data is successfully retrieved
                    retrieved_tickers.append(ticker)
                else:
                    print(f"Section with data-testid 'epsTrend' not found for {ticker}. Skipping ticker.")
            else:
                print(f"Failed to retrieve data for {ticker}. Status code: {response.status_code}")
        except Exception as e:
            print(f"An error occurred while fetching data for {ticker}: {e}")

    if headers is not None:
        # Create a pandas DataFrame
        df = pd.DataFrame(all_data, columns=headers)

        # Filter rows where 'CURRENCY IN USD' column is 'Current Estimate'
        df = df[df['CURRENCY IN USD'] == 'Current Estimate']

        # Add a new column 'Ticker' with ticker values corresponding to retrieved data
        df['Ticker'] = retrieved_tickers

        # Reset index
        df.reset_index(drop=True, inplace=True)

        return df
    else:
        print("No data found.")
        return None

# Example usage:
#tickers = ["AAPL", "GOOGL", "GOL", "ZNH", "MSFT", "HA"]  # Example tickers
table_df = scrape_yahoo_finance_tables(tickers)
table_df
```
